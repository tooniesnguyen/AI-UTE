# 50 Startup

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = '50_Startups.csv'
data = pd.read_csv(url)
data.head()
```

```python
data.hist();
```

```python
data.corr()
```

```python
sns.heatmap(data.corr(), annot = True);
```

```python
sns.pairplot(data);
```

```python
from sklearn import linear_model

X = data.loc[:,['R&D Spend','Administration','Marketing Spend']]
y = data['Profit']

model = linear_model.LinearRegression()
model.fit(X,y)
print("H s intercept l: ", model.intercept_)
print('H s coeffiient l: ',model.coef_)
print('H s  chnh xc: ',model.score( X, y))
```
